# Data Visualization📊📉

## 5 star hotels in different states

```python
star_5=df[df["StarRating"]==5]
```

```python
star_5
```

```python
star5_viz = star_5.groupby('State').size().reset_index()
star5_viz.rename(columns={0:'Hotels'},inplace=True)
star5_viz
```

```python
plt.rcParams['figure.figsize'] = (30,15)
plt.title('5 Star Hotels in different States')
p = sns.barplot(x='State',y='Hotels',data=star5_viz)
p.bar_label(p.containers[0])
p.set_xticklabels(labels=star5_viz['State'],rotation=35)
sns.set(font_scale=1.25)
plt.show()
```

```python
#The above viz. shows us that Kerala🌴🛶 has the most no. of 5 star hotels(42).
```

# 5 Star Hotel Rooms in different States

```python
star5_r = star_5.groupby('State')['Total Rooms'].sum().reset_index()
star5_r.sort_values(by='Total Rooms',ascending=False,inplace=True)
star5_r.head()
```

```python
plt.rcParams['figure.figsize'] = (30,15)
p = sns.barplot(x='State',y='Total Rooms',data = star5_r)
p.bar_label(p.containers[0])
plt.title("5 Star Hotel Rooms in each state")
p.set_xticklabels(labels=star5_r["State"], rotation=35)
sns.set(font_scale=1.25)
plt.show()
```

```python
#This viz. shows that Maharashtra has the highest no. of rooms(8886). Kerala comes in the 6 position with 4109 rooms
```

According to our previous observation, we found that Kerala has the highest no. of hotels registered with the Ministry Of Tourism. So here we will be focusing the state of Kerala.

```python
df1=df.copy
```

```python
df_kerala = df[df['State']=='Kerala'].reset_index()
df_kerala = df_kerala.sort_values(by='City')
df_kerala.head()
```

```python
df_kerala['City'].unique()
```

```python
df_kerala['City'] = df_kerala['City'].replace(['alleppey'],'Alappuzha') #alleppey - Alappuzha
df_kerala['City'] = df_kerala['City'].replace(['Idduki'],'Idukki') # Idduki - Idukki
df_kerala['City'] = df_kerala['City'].replace(['IRITTY'],'Iritty') # IRITTY - Iritty
df_kerala['City'] = df_kerala['City'].replace(['Ernakulam'],'Kochi') # Ernakulam - Kochi
df_kerala['City'] = df_kerala['City'].replace(['KASARGOD'],'Kasaragod') # KASARGOD - Kasaragod
df_kerala['City'] = df_kerala['City'].replace(['Calicut'],'Kozhikode') # Calicut - Kozhikode
df_kerala['City'] = df_kerala['City'].replace(['Karunagapally'],'Kollam') # Calicut - Kozhikode
df_kerala['City'].unique()
```

```python
k_c_h = df_kerala.groupby('City').size().reset_index().rename(columns={0:'Hotels'})
k_c_h.sort_values(by='Hotels',inplace=True)
#k_c_h['Labels'] = k_c_h['City']+'\n'+ str(k_c_h['Hotels'])
k_c_h['Labels']=""
for x in range(0,len(k_c_h)):
    k_c_h['Labels'][x] = k_c_h['City'][x]+'\n'+str(k_c_h['Hotels'][x])
```

```python
squarify.plot(k_c_h['Hotels'], label=k_c_h['Labels'],color=sns.color_palette("Spectral", len(k_c_h)),ec='black',alpha = 0.9)
plt.axis("off")
plt.show()
```

This treemap shows Kochi(Ernakulam District) has the most number of hotels.

# Hotel rooms in each city

```python
k_c_hr = df_kerala.groupby('City')['Total Rooms'].sum().reset_index()
k_c_hr.sort_values(by='Total Rooms',inplace=True,ascending=False)
k_c_hr.head()
```

```python
p = sns.barplot(x='City',y='Total Rooms',data=k_c_hr)
p.bar_label(p.containers[0])
plt.title("Hotel Rooms in each city")
p.set_xticklabels(labels=k_c_hr["City"], rotation=35)
plt.xticks(rotation=90)
sns.set(font_scale=1.25)
plt.show()
```

```python
#Again Kochi has the most number of rooms(4471).
```

# Hotels based on Alcohol Preferences

```python
plt.rcParams['figure.figsize'] = (10,5)
df_kerala.groupby('Alcohol').size().plot(kind='pie', autopct='%1.0f%%',subplots=True)
```

Due to less amount of data, 55% is "Not Mentioned", but "With Alcohol" preferences can be one of the reasons why many hotels are situated in Kerala.

As Kerala is focused on Tourism we will be analysing 5 star hotels in different parts of the state.

```python
k_5s = df_kerala[df_kerala['StarRating']==5]
k_5s.sort_values(by='City')
k_5s = k_5s.groupby('City').size().reset_index().rename(columns={0:'Hotels'}).sort_values(by='Hotels',ascending=False)
k_5s.head()
```

```python
plt.rcParams['figure.figsize'] = (20,10)
p = sns.barplot(x='City',y='Hotels',data=k_5s)
p.bar_label(p.containers[0])
plt.title("5 star Hotels in each city")
p.set_xticklabels(labels=k_5s["City"], rotation=35)
sns.set(font_scale=1.25)
plt.show()
```

Kochi dominates the hospitality and tourism sector with 19 registered 5 star hotels.
